##Matplotlib
* 파이썬에서 가장 널리 사용되는 2D 시각화 라이브러리
* 데이터를 다양한 형태의 그래프로 시각화

* SVG, PDF, PNG 등 다양한 포맷으로 저장 가능

#어떤 차트를 선택해야 할까?

image7.png

1. 📊 비교 (Comparison)
* 서로 다른 항목들 간의 값을 비교할 때 사용
* Bar chart: 항목 간 값 비교에 가장 일반적 (예: 제품별 매출)
* Line chart: 시간에 따른 비교, 변화량 비교 (예: 연도별 판매 추이)
* Radar chart: 다차원 요소(성능 등)를 한 눈에 비교 (예: 선수 능력 비교)
2. 📈 추세 (Trend)
* 시간의 흐름에 따라 데이터가 어떻게 변하는지 보여줌
* Bar chart: 시간 단위별 데이터 비교 (예: 월별 방문자 수)
* Line chart: 연속적 추세 분석 (예: 주가 변화)
* Area chart: 누적 변화량을 시각적으로 강조 (예: 총량 추이)
3. 📉 분포 (Distribution)
* 데이터가 어떻게 분포되어 있는지를 시각화
* Histogram: 연속형 변수의 빈도 분포 (예: 점수 분포)
* Box plot: 최소, 최대, 중앙값, 사분위수 등 통계 요약
* Scatter plot: 개별 데이터의 퍼짐과 밀도 확인
4. 🔗 상관관계 (Relationship)
* 두 변수 간의 연관성, 패턴, 상호작용을 보여줌
* Scatter plot: 상관관계 여부 시각화 (예: 공부시간 vs 성적)
* Heatmap: 변수 간 상관관계를 색상으로 표현 (예: 피어슨 상관계수)
* Network: 객체 간 연결 관계 표현 (예: 소셜 네트워크)
5. 🧩 구성 (Composition, %)
* 전체에 대한 부분 비율 또는 구조를 보여줌
* Pie chart: 비율로 나눠진 조각 시각화
* Stacked bar chart: 여러 구성요소를 누적해서 표현
* Treemap: 면적을 이용한 구성 비율 표현
6. 🔄 흐름 (Flow)
* 데이터가 어디에서 어디로 이동하는지를 보여줌
* Network: 흐름과 관계를 함께 시각화
* Sankey: 데이터 흐름의 양을 강조 (예: 에너지 흐름)
* Stream graph: 시간에 따른 양의 흐름 시각화 (예: 유튜브 장르 트렌드)

7.  library 별 차트 지원 여부

## 📊요일별 손님수 파악하기
* 요일별 손님 수의 차이를 시각적으로 비교하여 가장 붐비는 요일을 쉽게 파악하기 위해 막대 그래프 사용

## 막대 그래프
* 범주형 데이터의 크기 비교에 적합
* x축: 비교하고자 하는 범주 (예: 국가, 품목명 등)
* y축: 수치 값 (예: 수출 금액, 거래 건수 등)
* 막대의 높이: 해당 범주의 값의 크기
* 색상 (선택): 특정 기준에 따라 색을 다르게 적용할 수 있음

```python
#요일별 손님수
order = tips["day"].value_counts().index
sns.countplot(y="day", data=tips, hue="day", palette="RdBu", order=order)
plt.title("요일별 손님 수")
plt.show()
```

##📊식사 인원 수에 따른 평균 총 금액
* 식사 인원 수 증가에 따른 총 금액의 연속적인 변화를 직관적으로 보여주기 위해 선그래프를 사용

### 선 그래프

* 시간 흐름에 따른 수치 변화를 선(꺾은선)으로 표현하는 그래프
* 선: 연속된 값의 추이
* 시간에 따른 변화 관찰
* 증가 또는 감소 추세 파악
* 선이 위로 올라가면 증가, 내려가면 감소
* 점 간 간격이 넓을수록 변동이 큼

| 문자    | 마커 형태                      | 설명         |
| ----- | -------------------------- | ---------- |
| `'o'` | ● 원(circle)                | 가장 기본적인 마커 |
| `'s'` | ■ 사각형(square)              |            |
| `'^'` | ▲ 위쪽 삼각형(triangle\_up)     |            |
| `'v'` | ▼ 아래쪽 삼각형(triangle\_down)  |            |
| `'>'` | ▶ 오른쪽 삼각형(triangle\_right) |            |
| `'<'` | ◀ 왼쪽 삼각형(triangle\_left)   |            |
| `'D'` | ◆ 다이아몬드(diamond)           |            |
| `'d'` | ◇ 작은 다이아몬드(thin diamond)   |            |
| `'p'` | ✦ 오각형(pentagon)            |            |
| `'h'` | ⬡ 육각형1(hexagon1)           |            |
| `'H'` | ⬢ 육각형2(hexagon2)           |            |
| `'*'` | ★ 별(star)                  | 강조할 때 유용   |
| `'x'` | × 엑스(x)                    | 교차 표시      |
| `'+'` | + 플러스(plus)                |            |
| `'.'` | • 점(dot)                   | 매우 작은 점    |
| `','` | , 픽셀(pixel)                | 거의 안 보이는 점 |


```python
# 식사 인원 수에 따른 평균 총금액
grouped = tips.groupby("size")["total_bill"].mean()
plt.plot(grouped.index, grouped.values, label="평균 총 금액")
plt.xlabel("인원 수")
plt.ylabel("평균 총 금액")
plt.legend()
plt.show()
```

##📊총금액과 팁의 관계

#### 산점도

##📊점심과 저녁 팁 비교
* 점심과 저녁 시간대의 팁 금액 분포를 중앙값과 사분위수는 물론, 전체 분포 형태와 밀집 구간까지 함께 시각화하여 정밀하게 비교하기 위해 바이올린 플롯을 사용

#### violin plot
*  Box Plot과 Kernel Density Plot이 결합된 시각화 방법
* 데이터 분포의 형태와 중심 경향, 이상치를 함께 표현
* Box Plot처럼 중앙값, 사분위수, 이상치를 보여줌
* Density Plot처럼 **데이터의 분포 형태(비대칭, 치우침 등)**를 시각적으로 표현

| 요소         | 설명                               |
| ---------- | -------------------------------- |
| **중앙선**    | 데이터의 중앙값 (Median) 위치             |
| **박스 영역**  | 1사분위수(Q1)와 3사분위수(Q3) 사이 (IQR)    |
| **길쭉한 곡선** | 커널 밀도 함수 (KDE)를 통해 그려진 분포 형태     |
| **넓은 부분**  | 해당 값 근처에 데이터가 밀집되어 있음 (빈도 높음)    |
| **좁은 부분**  | 해당 값 근처에 데이터가 드뭄 (빈도 낮음)         |
| **이상치 점들** | 일반적인 분포 범위에서 벗어난 값들 (옵션에 따라 표시됨) |


#### violin plot 해석 방법

① 분포의 모양 파악
대칭형: 데이터가 평균을 중심으로 고르게 분포되어 있음

비대칭(왼쪽 또는 오른쪽 치우침): 분포의 꼬리가 어느 한쪽으로 길게 늘어져 있음

멀티모달: 곡선이 여러 개의 봉우리를 가질 경우, 데이터에 여러 집단이 존재할 가능성

② 데이터의 중심
중앙선의 위치: 중앙값이 어디에 위치하는지 확인
→ 분포의 중간값이 데이터의 중심을 얼마나 잘 대표하는지 알 수 있음

③ 데이터의 분산
바이올린의 전체 폭: 데이터가 얼마나 퍼져 있는지 보여줌
→ 폭이 넓을수록 해당 범위에 많은 데이터가 몰려 있음

④ 이상치 여부
일부 violin plot은 이상치를 점 형태로 따로 표시
→ 박스 외부의 점들을 통해 데이터의 특이값 확인 가능

##📊시간대 별 팁 금액 비율

#### 파이 차트


*  전체에서 각 항목이 차지하는 비율을 원형으로 나눈 차트
* 전체 대비 구성 비율 파악
* 주요 품목의 상대적 중요도 확인
* 큰 조각이 가장 많은 비중
* 파이 차트는 최대 6~8개 항목만 추천 항목 추가
* seaborn은 pie chart를 지원하지 않음

```python
#시간대 별 팁 금액 비율

#팁 금액대 분류
import pandas as pd

def tip_group(tip):
    if tip <= 2:
        return '2달러 이하'
    elif tip <= 5:
        return '2~5달러'
    else:
        return '5달러 초과'

tips['tip_range'] = tips['tip'].apply(tip_group)

# 시간대별 파이차트 그리기
for time in ['Lunch', 'Dinner']:
    data = tips[tips['time'] == time]
    counts = data['tip_range'].value_counts().sort_index()

    plt.figure(figsize=(6, 6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(f"{time} 시간대 팁 금액대 비율")
    plt.tight_layout()
    plt.show()
```

####relplot

* 산점도(scatter plot)나 선형 그래프(line plot)를 생성

#### 사용예시
~~~
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='time',           # 점심/저녁 분할
    hue='smoker',         # 흡연 여부 색상 구분
    style='sex',          # 성별에 따라 마커 모양 구분
    size='size',          # 식사 인원 수에 따라 크기 조절
    kind='scatter'
)

~~~

####옵션 설명
| 파라미터         | 설명                                         |
| ------------ | ------------------------------------------ |
| `data`       | 사용할 데이터프레임                                 |
| `x`, `y`     | X축, Y축 변수 (수치형)                            |
| `hue`        | 그룹별 색상 구분 (범주형)                            |
| `style`      | 마커 모양 구분 (범주형)                             |
| `size`       | 점의 크기 구분 (범주형 또는 수치형)                      |
| `col`, `row` | 열 또는 행으로 서브플롯 분할                           |
| `kind`       | `'scatter'` 또는 `'line'` (기본값: `'scatter'`) |
| `height`     | 서브플롯의 높이 (단위: 인치)                          |
| `aspect`     | 서브플롯의 종횡비                                  |
