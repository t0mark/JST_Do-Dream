###0. 시각화를 위한 전처리

* categorical : pandas.Categorical은 범주형 데이터(categorical data)를 효율적으로 표현하기 위한 Pandas의 데이터 타입
* 정렬 순서 설정(예: 10-14세가 00-04세보다 뒤에 오도록 명시 가능), Seaborn 등 그래프에서 카테고리 순서 자동 정렬 방지를 위해 사용

```python
# 전처리: 모든암 + 남녀전체 필터링
df_filtered = df[(df['성별'] == '남녀전체') & (df['암종'] == '모든암')].copy()

# 연령군 정렬
age_order = ['00-04세', '05-09세', '10-14세', '15-19세', '20-24세',
             '25-29세', '30-34세', '35-39세', '40-44세', '45-49세',
             '50-54세', '55-59세', '60-64세', '65-69세', '70-74세',
             '75-79세', '80-84세', '85세이상']
#연령군으로 순서가 있는 범주형 생성
df_filtered['연령군'] = pd.Categorical(df_filtered['연령군'], categories=age_order, ordered=True)
df_filtered.head()
```

### 4. 특정 연도 기준 연령대별 암 발생자수 (누적 막대그래프)

* pivot_table : Pandas에서 데이터를 요약 정리하는 도구
* 행(row)과 열(column)을 기준으로 데이터를 집계(aggregation)
~~~
DataFrame.pivot_table(
    data,              # 데이터프레임
    index=None,        # 행 인덱스로 사용할 열
    columns=None,      # 열 인덱스로 사용할 열
    values=None,       # 집계할 값
    aggfunc='mean',    # 집계 함수 (sum, mean, count 등)
    fill_value=None    # NaN을 채울 값
)
~~~

```python
# 특정 연도 기준 연령대별 암 발생자수
pivot = df_filtered[df_filtered['발생연도'] == 2019].pivot_table(
    index='연령군',
    columns='성별',
    values='발생자수',
    aggfunc='sum').fillna(0)

#---------- 차트 생성  코드 여기에 ----------------

#---------- 차트 생성  코드 여기에 ----------------
plt.title('2019년 연령대별 암 발생자수')
plt.xlabel('연령군')
plt.ylabel('발생자수')
plt.legend(title='성별')
plt.tight_layout()
plt.show()
```
